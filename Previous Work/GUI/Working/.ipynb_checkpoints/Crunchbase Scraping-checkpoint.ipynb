## Incorporating the code


```python
investor = []
investor_type = []
number_of_investments = []
number_of_exits = []
location = []
description = []
website = []
number = []
email = []
founders = []
```

```python
for i in range(65): 
    if i%10 == 0:
        time.sleep(15)
        print("10 pages waiting initiated")
    
    print("Page Number " + str(i))
    
    rows = wd.find_elements_by_css_selector('grid-row.ng-star-inserted')
    print("Number of results on the page " + str(len(rows)))
    
    if len(rows)< 50:
        print("trying again")
        time.sleep(4)
        rows = wd.find_elements_by_css_selector('grid-row.ng-star-inserted')
        print("New Number of results on the page " + str(len(rows)))
        
    for row in rows:
        data = row.text.split('\n')
        investor.append(data[1])
        investor_type.append(data[2])
        number_of_investments.append(data[3])
        number_of_exits.append(data[4])
        location.append(data[5])
        description.append(data[6])
        website.append(data[7])
        number.append(data[8])
        if (len(data)) < 11:
            data.append("ERROR FOUND - CHECK MANUALLY")
            if len(data) < 10:
                data[9].append("ERROR FOUND - CHECK MANUALLY")
        email.append(data[9])
        founders.append(data[10])
    
    next_b = wd.find_element_by_xpath("//icon[@key='icon_caret_right']")
    wd.execute_script("arguments[0].click();", next_b)
    time.sleep(5)
    print("Page should be loaded by now!")
```

```python
import csv
spamWriter = csv.writer(open('Accelerators_crunchbase.csv', 'a'))
for i in range(len(investor))[3234:]:
    spamWriter.writerow([investor[i], investor_type[i], number_of_investments[i], number_of_exits[i], location[i], description[i], website[i], number[i], email[i], founders[i]])
```
